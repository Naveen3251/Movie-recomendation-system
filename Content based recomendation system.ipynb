```python
#the above o/p shows that genres column has list of dictionaris in form string hence we going to choose ast module for process
```

```python
def ast_convert(data):
    lis=[]
    for d in ast.literal_eval(data):
            lis.append(d['name'])
    return lis
   
```

```python
def cast_convert(data):
    lis=[]
    count=0
    for i in ast.literal_eval(data):
        if count>=3:
            break
        lis.append(i['name'])
        count+=1
    return lis
data['cast']=data['cast'].apply(cast_convert)
        
```

```python
def director(data):
    l=[]
    for d in ast.literal_eval(data):
        if d['job']=='Director':
            l.append(d['name'])
    return l
```

# Preprocessing the text

# 1]Change to lower case

```python
data_new['tags']=data_new['tags'].apply(lambda x:x.lower())
        
```

```python
data_new['tags']
```

```python
stopwords=STOP_WORDS
nlp=spacy.load('en_core_web_sm')
```

# 2.Tokenization and Removing Punctuations

```python
def remove_pun(text):
    doc=nlp(text)
    rem=[pun for pun in doc if not pun.is_punct]
    return " ".join(token.text for token in rem)
```

```python
data_new['tags']=data_new['tags'].apply(remove_pun)
```

```python
data_new['tags'][0]
```

# 3.Lemmatization and Removing Stopwords

```python
def remove_stop(text):
    doc=nlp(text)
    lem=[w.lemma_ for w in doc]
    st=[stop for stop in lem if stop not in stopwords]
    return " ".join(l for l in st)
```

```python
data_new['tags']=data_new['tags'].apply(remove_stop)
```

```python
data_new['tags'][0]
```

# Cosine similarities

```python
# our model should be capable of finding the similarity between movies based o their tags
# Our Recommender model takes a movie title as input and predicts top-n most similar movies based on tags
# here we will use concept of Cosine distance to calculate the similarity of tags
```

```python
#cosine similarity

#Formula
#Cos(x,y)=x.y/||x|| *||y||
#x.y-->dot product of two vectors
#||x||,||y|| -->length of two vectors 'x' and 'y'
#||x|| *||y||-->cross product of two vectors 'x' and 'y'
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vector).shape
```

```python
similarity=cosine_similarity(vector)
print(similarity[0])
```

# Function for movie recommendation system

```python
#defining the function
def recommend(movie):
    try:
            #to get the movie index
            movie_index=data_new[data_new['title']==movie].index[0]
            #distances
            distance=similarity[movie_index]
            movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
            for movies in movie_list:
                 print(data_new.iloc[movies[0]].title)
    except:
        print('Enter the movie name correctly')
    
```

```python
recommend('Avatar')
```

```python
recommend('The Dark Kight Rises')
```

```python
recommend('Spectre')
```
